Statistical Analysis with NumPy

```python
mean_rating = df_cleaned['vote_average'].mean()
median_rating = df_cleaned['vote_average'].median()
std_rating = df_cleaned['vote_average'].std()

print(f"Mean rating: {mean_rating:.2f}")
print(f"Median rating: {median_rating:.2f}")
print(f"Standard deviation of ratings: {std_rating:.2f}")
```

```python
mean_budget = df_cleaned['budget'].mean()
median_budget = df_cleaned['budget'].median()
std_budget = df_cleaned['budget'].std()

print(f"Mean budget: {mean_budget:.2f}")
print(f"Median budget: {median_budget:.2f}")
print(f"Standard deviation of budget: {std_budget:.2f}")

mean_revenue = df_cleaned['revenue'].mean()
median_revenue = df_cleaned['revenue'].median()
std_revenue = df_cleaned['revenue'].std()

print(f"\nMean revenue: {mean_revenue:.2f}")
print(f"Median revenue: {median_revenue:.2f}")
print(f"Standard deviation of revenue: {std_revenue:.2f}")
```

Grouped Analysis with Pandas

```python
genre_ratings = df_cleaned.groupby('genres')['vote_average'].mean().sort_values(ascending=False)
display(genre_ratings)
```

```python
director_ratings = df_cleaned.groupby('director')['vote_average'].mean().sort_values(ascending=False)
display(director_ratings)
```

```python
df_cleaned['release_date'] = pd.to_datetime(df_cleaned['release_date'])
df_cleaned['release_year'] = df_cleaned['release_date'].dt.year

yearly_ratings = df_cleaned.groupby('release_year')['vote_average'].mean()
display(yearly_ratings)
```

Visualizations with Matplotlib

```python
import matplotlib.pyplot as plt

# Plot average rating by genre (top 10)

genre_ratings.head(10).plot(kind='bar')
plt.title('Top 10 Genres by Average Rating')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot average rating by director (top 10)

director_ratings.head(10).plot(kind='bar')
plt.title('Top 10 Directors by Average Rating')
plt.xlabel('Director')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

yearly_ratings.plot(kind='line')
plt.title('Average Movie Rating Over Time')
plt.xlabel('Release Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.scatter(df_cleaned['budget'], df_cleaned['vote_average'], alpha=0.5)
plt.title('Budget vs. Average Rating')
plt.xlabel('Budget')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.scatter(df_cleaned['revenue'], df_cleaned['vote_average'], alpha=0.5)
plt.title('Revenue vs. Average Rating')
plt.xlabel('Revenue')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_rated_movies = df_cleaned.sort_values('vote_average', ascending=False).head(10)


plt.barh(top_rated_movies['title'], top_rated_movies['vote_average'], color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies by Average Rating')
plt.gca().invert_yaxis()  # To display the highest rating at the top
plt.tight_layout()
plt.show()
```
